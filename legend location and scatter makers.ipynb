# plt legend location: 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 10, 0.1)
s = np.sin(x)
c = np.cos(x)
```

### Place the anchor at the top right and align the top right of the legend frame with the anchor. Don't leave any blank spaces.

```python
plt.plot(x, s, "r", label="sin")
plt.plot(x, c, "k", label="cos")
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=18)
```

### Place the anchor at the top right and align the top right of the legend frame with the anchor. Leave some space.

```python
plt.plot(x, s, "r", label="sin")
plt.plot(x, c, "k", label="cos")
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1, fontsize=18)
```

### Place the anchor at the bottom right and align the bottom right of the legend frame with the anchor. Leave some space.

```python
plt.plot(x, s, "r", label="sin")
plt.plot(x, c, "k", label="cos")
plt.legend(bbox_to_anchor=(1, 0), loc='lower right', borderaxespad=1, fontsize=18)
```

### Place the anchor outside at the bottom left, and align the top left of the legend frame with the anchor.

```python
plt.plot(x, s, "r", label="sin")
plt.plot(x, c, "k", label="cos")
plt.legend(bbox_to_anchor=(0, -0.1), loc='upper left', borderaxespad=0, fontsize=18)
```

### Place the anchor at the outer upper right corner and align the upper left corner of the legend frame with the anchor.

```python
plt.plot(x, s, "r", label="sin")
plt.plot(x, c, "k", label="cos")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=18)
```

# plt.scatter makers:

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(1, 11)
y1 = np.repeat(3, 10)
y2 = np.repeat(2, 10)
y3 = np.repeat(1, 10)
 
markers1 = [".", ",", "o", "v", "^", "<", ">", "1", "2", "3"]
markers2 = ["4", "8", "s", "p", "*", "h", "H", "+", "x", "D"]
markers3 = ["d", "|", "_", "None", None, "", "$x$",
            "$\\alpha$", "$\\beta$", "$\\gamma$"]
for i in x-1:
  plt.scatter(x[i], y1[i], s=300, marker=markers1[i])
  plt.scatter(x[i], y2[i], s=300, marker=markers2[i])
  plt.scatter(x[i], y3[i], s=300, marker=markers3[i])
```
